```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('code')
import models

np.random.seed(4)

fig = plt.figure(figsize = (10, 8))
plot_s = 50000
plot_f = 100000

for i in range(2):
    # Set the strength of recurrent excitation W_EE
    w_ee_set = [3.5, 3.75][i] 

    # Create an instance of the excitatory-inhibitory network (EIN) model with the specified W_EE value
    test = models.LocalEI(w_ee = w_ee_set, total_t=100000)
    test.simulate()  # simulate the slowwave model

    # Plot time series of the excitatory/inhibitroy firing rates and adaptation
    time_series = np.arange(0, 5000, 0.1)
    ax = fig.add_subplot(3, 2, 1 + i)
    ax.set_title('W_EE = {}'.format(w_ee_set))
    ax.plot(time_series, test.r_e[plot_s : plot_f], color='orange')
    ax.set_ylim(-0.3, 13.0)
    ax.set_ylabel('Excicatory')
    ax = fig.add_subplot(3, 2, 3 + i)
    ax.plot(time_series, test.r_i[plot_s : plot_f], color='navy')
    ax.set_ylim(-0.3, 40.0)
    ax.set_ylabel('Inhibitory')
    ax = fig.add_subplot(3, 2, 5 + i)
    ax.plot(time_series, test.adapt[plot_s : plot_f])
    ax.set_ylim(0, 2.2)
    ax.set_xlabel('Time [ms]')
    ax.set_ylabel('Adaptation')
    
plt.show()
```
